## Предобработка данных

Изменим название столбца с расстоянием до центра города, приведем его к "змеиному" регистру:

```python
df.columns = ['total_images', 'last_price', 'total_area', 'first_day_exposition',
       'rooms', 'ceiling_height', 'floors_total', 'living_area', 'floor',
       'is_apartment', 'studio', 'open_plan', 'kitchen_area', 'balcony',
       'locality_name', 'airports_nearest', 'city_centers_nearest',
       'parks_around3000', 'parks_nearest', 'ponds_around3000',
       'ponds_nearest', 'days_exposition']
```

Далее нужно решить в каких данных мы сможем заполнить пропуски, а в каких нет.
Заполнить пропуски мы сможем в столбцах:
- balcony, если значение NaN будем заменять его на 0, потому что если нет иного значения, значит и балкона скорее всего нет;
- ceiling_height, предположим, что пропущенные значения можно заменить средним значением.
Остальные значения возможно придется так и оставить с пропусками, потому что их невозможно корректно заполнить, не повлияв на результаты анализа.

Возможно есть смысл заполнить пропущенные значения в столбце с площадью кухни пропорционально жилой площади. 
Проверим гипотезу, что чем больше площадь жилого помещения, тем больше площадь кухни. Для этого проверим есть ли корреляция между этими двумя столбцами.

```python
df['living_area'].corr(df['kitchen_area'])
```

Можно сказать, что взаимосвязь есть, но не сильная, поэтому примем решение не заполнять столбец с площадью кухни пропорционально площади жилья. Гипотеза не подтвердилась.

На пропуски в данных могли повлиять следующие факторы:
1) технический;
 - часть массива получена автоматически на основе картографических данных, вполне возможно, что имеет место техническая ошибка: некорректно указана улица или населенный пункт (с опечаткой), сбой в работе системы сбора данных и т.д.
 
2) человеческий;
 - остальная часть данных вписана собственноручно пользователем, опечатка или просто невнимательность, забывчивость и вот уже пропуски в данных.
 
Если бы это был рабочий проект, сообщила бы заказчику об отсутствии большого количества данных.

Заполним пропуски в столбце balcony:

```python
df['balcony'] = df['balcony'].fillna(0)
```

Проверим, что пропуски заполнены:

```python
df['balcony'].isna().sum()
```

Заполним пропуски в столбце с высотой потолка. Вызовем метод describe() и посмотрим на значения в этом столбце:

```python
df['ceiling_height'].describe()
```

В данных есть помещение с высотой потолков в 100 метров. Посмотрим, что это за строки

```python
df.query('ceiling_height == 100')
```

Одна строка, ни одной фотографии, 1 комната, маленькая площадь. Очень похоже на опечатку.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
```

Проверим, что пропусков больше нет

```python
df['ceiling_height'].isna().sum()
```

Заполним пропуски в столбце locality_name, среднего здесь нет и не может быть, поэтому заполним пропуски "Неизвестно"

```python
df['locality_name'] = df['locality_name'].fillna('Неизвестно')
```

```python
df['locality_name'].isna().sum()
```

Заменим пропуски в floors_total:

```python
df['floors_total'] = df['floors_total'].fillna(0)
```

```python
df['floors_total'].isna().sum()
```

## Расчёты и добавление результатов в таблицу

Необходимо посчитать и добавить в таблицу столбец с ценой квадратного метра:

```python
df['metr_cost'] = df['last_price']/df['total_area']
```

```python
df.head(5)
```

Добавим в таблицу день недели, месяц и год публикации объявления:

```python
df['w_publ'] = df['first_day_exposition'].dt.weekday
df['m_publ'] = df['first_day_exposition'].dt.month
df['y_publ'] = df['first_day_exposition'].dt.year
```

```python
df.head(5)
```

Добавим столбец с варинтом этажа: "первый", "последний","другой", для этого напишем функцию:

```python
df.info()
```

```python
def floor_category(floor,floors_total):
    if floor <= floors_total:
        if floor == 1:
            return 'первый'
        elif floor == floors_total:
            return 'последний'
        return 'другой'
    return 'неизвестно'
    
```

Проверим, что функция работает корректно:

```python
display(floor_category(1,2))
display(floor_category(2,2))
display(floor_category(2,4))
display(floor_category(10,4))
```

Применим функцию к двум столбцам и добавим в таблицу новый столбец с категорией этажа:

```python
df['floor_categories'] = df.apply(lambda x: floor_category(x['floor'],x['floors_total']), axis =1)
```

Проверим какие теперь значения в столбце floor_categories:

```python
df['floor_categories'].value_counts()
```

Проверим отношение общей площади к жилой и кухне:

```python
df['living_total_ratio'] = df['living_area']/df['total_area']*100
df['kitchen_total_ratio'] = df['kitchen_area']/df['total_area']*100
```

```python
df.head(5)
```

###  Факторы, влияющие на стоимость квартиры

```python
display(df['last_price'].corr(df['total_area']))
display(df['last_price'].corr(df['rooms']))
display(df['last_price'].corr(df['city_centers_nearest']))
```

Можно сказать, что:
- цена напрямую зависит от площади жилья,

Изучим зависимость цены от этажа. Для этого создадим новый столбец с id этажа ( первый - 1, последний - 2, другой - 0)

```python
def floor_categories_id(floor_categories):
    if floor_categories == 'первый':
        return 1
    elif floor_categories == 'последний':
        return 2
    elif floor_categories == 'другой' and 'неизвестно':
        return 0
    else:
        return 0
```

```python
df['floor_categories_id'] = df['floor_categories'].apply(floor_categories_id)
```
